```
url = "https://en.wikipedia.org/wiki/List_of_companies_of_Pakistan"
```

```
headers = {"User-Agent": "Mozilla/5.0"}
```

```
response = requests.get(url, headers=headers)
```

```
soup = BeautifulSoup(response.text, "html.parser")
```

```
tables = soup.find_all("table", class_="wikitable")
```

```
records = []
```

```
for table in tables:
    rows = table.find_all("tr")[1:]  # skip header
    for row in rows:
        cols = [c.text.strip() for c in row.find_all("td")]

        # We deliberately ignore "Notable firms"
        records.append({
            "Company_Name": cols[0] if len(cols) > 0 else None,
            "Industry": cols[1] if len(cols) > 1 else None,
            "Sector": cols[2] if len(cols) > 2 else None,
            "Headquarters": cols[3] if len(cols) > 3 else None,
            "Founded_Year": cols[4] if len(cols) > 4 else None
        })
```
