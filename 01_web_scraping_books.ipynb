```python
!pip install lxml bs4
import pandas as pd
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
url = "http://books.toscrape.com/catalogue/page-1.html"
response = requests.get(url)
html = response.text
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
books = soup.find_all("article", class_="product_pod")
len(books)
```

```python
book = books[0]

title = book.h3.a["title"]
price = book.find("p", class_="price_color").text
rating = book.find("p", class_="star-rating")["class"][1]
availability = book.find("p", class_="instock availability").text.strip()

title, price, rating, availability
```

```python
data = []

for book in books:
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text
    rating = book.find("p", class_="star-rating")["class"][1]
    availability = book.find("p", class_="instock availability").text.strip()

    data.append([title, price, rating, availability])
```

```python
df = pd.DataFrame(data, columns=["Title", "Price", "Rating", "Availability"])
df.head()
```

```python
all_data = []

for page in range(1, 6):  # first 5 pages for demo
    url = f"http://books.toscrape.com/catalogue/page-{page}.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "lxml")
    books = soup.find_all("article", class_="product_pod")

    for book in books:
        title = book.h3.a["title"]
        price = book.find("p", class_="price_color").text
        rating = book.find("p", class_="star-rating")["class"][1]
        availability = book.find("p", class_="instock availability").text.strip()

        all_data.append([title, price, rating, availability])
```

```python
final_df = pd.DataFrame(all_data, columns=["Title", "Price", "Rating", "Availability"])
final_df.shape
```

```python
final_df.to_csv("books_data.csv", index=False)
```
