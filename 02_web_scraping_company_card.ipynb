```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://webscraper.io/test-sites/e-commerce/static/computers/laptops"
response = requests.get(url)
soup = BeautifulSoup(response.text, "lxml")
```

```python
products = soup.find_all("div", class_="thumbnail")
len(products)
```

```python
p = products[0]

name = p.find("a", class_="title").text.strip()
price = p.find("h4", class_="price").text.strip()
description = p.find("p", class_="description").text.strip()

reviews = p.find("div", class_="ratings").find("p").text.strip()
rating = len(p.find("div", class_="ratings").find_all("span", class_="glyphicon-star"))

name, price, description, reviews, rating
```

```python
data = []

for p in products:
    name = p.find("a", class_="title").text.strip()
    price = p.find("h4", class_="price").text.strip()
    description = p.find("p", class_="description").text.strip()

    reviews = p.find("div", class_="ratings").find("p").text.strip()
    rating = len(p.find("div", class_="ratings").find_all("span", class_="glyphicon-star"))

    data.append([name, price, description, reviews, rating])
```

```python
df = pd.DataFrame(data, columns=["Name", "Price", "Description", "Reviews", "Ratings"])
df.head()
```

```python
df.to_csv("laptops_data.csv", index=False)
```
